1.หาพื้นที่ (district_name)ที่มีจำนวนโครงการเยอะที่สุดและมีความต้องการเช่าเยอะที่สุด (5 คะแนน)

```python
dataall = project1.merge(rent1, left_on='row_number', right_on='row_number')
dataall.head()
```

```python
dataall2 =dataall.merge(eat1, left_on='row_number', right_on='row_number')
dataall2.describe()
```

```python
dataall2.groupby("district_name_th_y")[["name_th"]].max().sort_index(ascending=False) #_y = ชื่อพื้นที่ทั่วประเทศ ทีมีโครงการเยอะสุด
ans_1_1 = dataall2.groupby("district_name_th_y")[["name_th"]].max().reset_index().sort_index(ascending=False) 
ans_1_1.iloc[[0]]
```

```python
dataall2.groupby("volume_of_rental")[["district_name_th_y"]].max().sort_index(ascending=False) #_y = ชื่อพื้นที่ทั่วประเทศ ทีมีโครงการเยอะสุด
ans_1_2 = dataall2.groupby("volume_of_rental")[["district_name_th_y"]].max().sort_index(ascending=False) 
ans_1_2.iloc[[0]]
```

```python
print(f'พื้นที่ที่มีโครงการเยอะที่สุด คือ {ans_1_1.iloc[0,0]} โครงการที่เยอะที่สุดคือ {ans_1_1.iloc[0,1]}')
print(f'แต่พื้นที่ที่มีความต้องการเช่ามากที่สุด คือ {ans_1_2.iloc[0,0]}')
```
